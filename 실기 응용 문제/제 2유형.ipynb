# 제 2유형
고객 3,500명에 대한 학습용 데이터(y_train.csv, X_train.csv)를 이용하여 성별예측 모형을 만든
후, 이를 평가용 데이터(X_test.csv)에 적용하여 얻은 2,482명 고객의 성별 예측값(남자일 확률)을
다음과 같은 형식의 CSV 파일로 생성하시오.(제출한 모델의 성능은 ROC-AUC 평가지표에 따라
채점)

```python
X_train=pd.read_csv('data/X_train.csv', encoding='cp949')
X_test=pd.read_csv('data/X_test.csv', encoding='cp949')
y_train=pd.read_csv('data/y_train.csv')

col_1_name='cust_id'
col_2_name='gender'

col_1=X_test.cust_id
```

```python
X_train.index=X_train.cust_id
X_train.drop('cust_id', axis=1, inplace=True)

X_test.index=X_test.cust_id
X_test.drop('cust_id', axis=1, inplace=True)
```

```python
y_train.index=y_train.cust_id
y_train.drop('cust_id', axis=1, inplace=True)
```

```python
data=pd.concat([X_train, y_train], axis=1)
data.isnull().sum()
```

```python
X_train.isnull().sum()
```

```python
data.fillna(0, inplace=True)
X_test.fillna(0, inplace=True)
```

```python
data.describe()
```

```python
data
```

```python
data=data[data.총구매액>=0]
data.corr()
```

```python
data.describe()
```

```python
X_train=data[data.columns[:-1]]
y_train=data[['gender']]

X_train_dum=pd.get_dummies(X_train)
X_test_dum=pd.get_dummies(X_test)
```

```python
missing_cols=set(X_train_dum.columns)-set(X_test_dum.columns)
for col in missing_cols:
    X_test_dum[col]=0

remain_cols=set(X_test_dum.columns)-set(X_train_dum.columns)
X_test_dum.drop(remain_cols, axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid=train_test_split(X_train_dum, y_train, stratify=y_train, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X_train)
X_scaled_train=pd.DataFrame(scaler.transform(X_train), columns=X_train.columns)
X_scaled_valid=pd.DataFrame(scaler.transform(X_valid), columns=X_train.columns)
X_scaled_test=pd.DataFrame(scaler.transform(X_test_dum), columns=X_test_dum.columns)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(C=0.0001,solver='newton-cg')
model.fit(X_scaled_train, y_train)
pred_train=model.predict(X_scaled_train)
model.score(X_scaled_train, y_train)
```

```python
pred_valid=model.predict(X_scaled_valid)
model.score(X_scaled_valid, y_valid)
```

```python
from sklearn.model_selection import GridSearchCV
grid_search=GridSearchCV(LogisticRegression(),
                        param_grid={'C':[0.0001,0.001,0.01,0.1,1,10,100,1000],
                                    'solver':['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']},
                        cv=5)
grid_search.fit(X_scaled_train, y_train)
grid_search.best_params_
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_valid, model.decision_function(X_scaled_valid))
```

```python
col_2=model.predict_proba(X_scaled_test)[:,0]
```
