# Question 1
mtcars 데이터셋(mtcars.csv)의 qsec 컬럼을 최소최대 척도(Min-Max Scale)로 변환한 후 0.5보다
큰 값을 가지는 레코드 수를 구하시오.

```python
import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')

import pandas as pd
data=pd.read_csv('data/mtcars.csv', encoding='utf-8', index_col=0)
qsec=data[['qsec']]

data
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(qsec)
qsec_scaled=scaler.transform(qsec)
qsec_scaled
```

```python
print(len(qsec_scaled[qsec_scaled>0.5]))
```

# Question 3
mtcars 데이터셋(mtcars.csv)의 wt 컬럼의 이상치(IQR 1.5 외부에 존재하는)값들을 outlier 변수에 저장하라

```python
wt=data.wt
```

```python
Q75=wt.quantile(0.75)
Q25=wt.quantile(0.25)
IQR=Q75-Q25
IQR
```

```python
import numpy as np
outlier=np.array(wt[(wt<Q25-IQR*1.5)|(wt>Q75+IQR*1.5)])
```
