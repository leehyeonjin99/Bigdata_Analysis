# 06
telco-customer-churn 데이터 세트이다. TotalCharges 항목에서 이상값을 제외한 평균을 구하시오.

```python
data=pd.read_csv('data/WA_Fn-UseC_-Telco-Customer-Churn.csv')
data.head()
```

```python
data.TotalCharges=pd.to_numeric(data.TotalCharges.replace(' ',''))
```

```python
mean=data.TotalCharges.mean()
std=data.TotalCharges.std()
```

```python
result=data[data.TotalCharges<mean+1.5*std].TotalCharges.mean()
print(result)
```

# 18
다음은 marvel 데이터 세트이다. Hair가 "Brown Hair"이고 Eye가 "Brown Eyes"인 데이터를 훈련 데이터로 추출했을 때, APPEARANCES에서 이상값을 제외한 평균을 구하시오.

```python
data=pd.read_csv('data/marvel-wikia-data.csv')
data.head()
```

```python
train_data=data[(data.HAIR=='Brown Hair')&(data.EYE=='Brown Eyes')]
train_data.head()
```

```python
mean=train_data.APPEARANCES.mean()
std=train_data.APPEARANCES.std()
```

```python
result=train_data[(train_data.APPEARANCES<mean+1.5*std)&(train_data.APPEARANCES>mean-1.5*std)].APPEARANCES.mean()
print(result)
```
