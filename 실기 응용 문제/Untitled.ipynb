#### mtcars 데이터셋(data/mtcars.csv)의 qsec 컬럼을 최소최대 척도(Min-Max Scale)로 변환한 후 0.5보다 큰 값을 가지는 레코드 수를 구하시오.


```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np

data=pd.read_csv('data/mtcars.csv', index_col=0)
data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(data[['qsec']])
data.qsec=scaler.transform(data[['qsec']])
```

```python
len(data[data.qsec>=0.5])
```

#### 데이터를 StandardScaler을 한 다음, 2차원으로 PCA를 이용해서 차원축소 해라.

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np

data=pd.read_csv('data/mtcars.csv', index_col=0)

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_data=pd.DataFrame(scaler.fit_transform(data))

scaled_data.head()
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca_2=pca.fit_transform(scaled_data)
pca_data=pd.DataFrame(pca_2, columns=['pca_col1','pca_col2'])
pca_data.head()
```

#### KMeans를 이용하여 데이터들을 군집 3개로 분류하고 평가하라. 

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np

data=pd.read_csv('data/mtcars.csv', index_col=0)

from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3, random_state=0)
kmeans.fit(data)
data['cluster']=kmeans.labels_
data.head()
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
samples_score=silhouette_samples(data.drop('cluster', axis=1), data.cluster)
data['silhouette_score']=samples_score
data.head()
```

```python
average_score=silhouette_score(data.drop('cluster', axis=1),data.cluster)
print("모든 데이터의 평균 실루엣 계수 :",average_score.round(3))
```

```python
# 군집별 평균 실루엣 계수
data.groupby('cluster')['silhouette_score'].mean()
```

#### 신뢰구간, 카이제곱 구하기

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np

data=pd.read_csv('data/mtcars.csv', index_col=0)
```

```python
from scipy import stats
interval=stats.t.interval(alpha=0.95, df=len(data.mpg)-1, loc=data.mpg.mean(), scale=np.std(data.mpg, ddof=1)/np.sqrt(len(data.mpg)))
print(interval)
```

```python
print("카이스퀘어 통계량 :",stats.chi2_contingency(data)[0])
```

#### 판다스의 to_datetime을 이용해 시계열 컬럼을 생성하고 원하는 시간을 추출하자 타임 델타도 계산해보자

```python
date=[]
for i in range(1,33):
    if i<=9:
        date.append('2021-12-0'+str(i))
    elif i<=31:
        date.append('2021-12-'+str(i))
    else:
        date.append('2022-01-0'+str(i-31))
date
```

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np

data=pd.read_csv('data/mtcars.csv', index_col=0)
data['date']=pd.to_datetime(date, format='%Y-%m-%d')
data.info()
```

```python
data['year']=data.date.dt.year
data['month']=data.date.dt.month
data['day']=data.date.dt.day
data['date_m']=data.date.dt.to_period(freq='M')
data.tail()
```

```python
past=pd.to_datetime('2020-06-18')
data['time_delta']=data.date-past
data.head()
```
