주어진 데이터 첫 번째 행부터 순서대로 80%까지의 데이터를 훈련 데이터로 추출 후 housing_office 항목에서 'total_bedrooms' 변수의 결측값을 'total_bedrooms' 변수의 중앙값으로 대체하고 대체 전의 'total_bedrooms' 변수 표준편차값을 산출하려 한다. 결측값을 중앙값으로 변환한 후, 변환 이전과 이후의 표준편차 차이를 구하시오.

```python
data=pd.read_csv('data/housing.csv')
data.head()
```

```python
train=data[:int(data.shape[0]*0.8)]
test=data[int(data.shape[0]*0.8):]
```

```python
before_std=train.total_bedrooms.std()
print(before_std)
```

```python
median=train.total_bedrooms.median()
train.total_bedrooms.fillna(median, inplace=True)
after_std=train.total_bedrooms.std()
print(after_std)
```

```python
result=np.abs(after_std-before_std)
print(result)
```

# 03
Charges 항목에서 이상값의 합을 구하시오. (이상값은 평균에서 1.5 표준편차 이상인 값)

```python
data=pd.read_csv('data/insurance.csv')
data.head()
```

```python
mean=data.charges.mean()
mean
```

```python
std=data.charges.std()
std
```

```python
high=mean+1.5*std
low=mean-1.5*std
```

```python
result=data[(data.charges>=high)|(data.charges<=low)]['charges'].sum()
print(result)
```

기업에서 생성된 주문데이터이다. 80,009건의 데이터에 대하여 정시도착 가능여부 예측 모델을 만들고, 평가 데이터에 대하여 정시도착 가능여부 예측 확률을 기록한 CSV를 생성하시오.

```python
data=pd.read_csv('data/Train.csv')
data
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
X=pd.get_dummies(data[data.columns[1:-1]])
y=data[data.columns[-1]]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, stratify=y)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X_train)
X_scaled_train=scaler.transform(X_train)
X_scaled_test=scaler.transform(X_test)
```

```python
# LogisticRegression
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_scaled_train, y_train)
model.score(X_scaled_train, y_train)
```

```python
model.score(X_scaled_test, y_test)
```

```python
# Support Vector Machine
from sklearn.svm import SVC
model=SVC()
model.fit(X_scaled_train, y_train)
model.score(X_scaled_train, y_train)
```

```python
model.score(X_scaled_test, y_test)
```
