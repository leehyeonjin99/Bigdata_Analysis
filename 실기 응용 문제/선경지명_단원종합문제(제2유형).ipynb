# 01
주어진 데이터는 각 고객이 가입한 서비스와 계정 정보, 인구에 대한 통계 정보들이가. 주어진 훈련 데이터를 이용하여 모델을 훈련한 후 테스트 데이터로 고객의 이탈 여부를 예측하고 csv 포맷으로 제출하시오. (단, 이탈 : "Yes", 유지 : "No")

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np

data=pd.read_csv('data/WA_Fn-UseC_-Telco-Customer-Churn.csv')
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
X=data[data.columns[:-1]]
y=data[['Churn']]
```

```python
X_dummies=pd.get_dummies(X)
X_dummies
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X_dummies, y, stratify=y, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X_train)
X_scaled_train=scaler.transform(X_train)
X_scaled_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
#model=LogisticRegression()
model=DecisionTreeClassifier(max_depth=9)
#model=RandomForestClassifier(max_depth=10)
model.fit(X_scaled_train, y_train)
model.score(X_scaled_train, y_train)
```

```python
model.score(X_scaled_test, y_test)
```

```python
'''from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, model.dec'''
```

```python
pred_test=model.predict(X_scaled_test)
pred_prob=model.predict_proba(X_scaled_test)
```

```python
y_test['pred']=pred_test
y_test[['prob_No','prob_Yes']]=pred_prob
y_test
```

```python
y_test.to_csv('data/Churn_Predict.csv')
```

# 02
다음은 mtcars 데이터 세트로 32개 자동차들의 디자인과 성능을 비교한 데이터이다. 훈련 데이터와 평가 데이터를 7:3으로 분할한 후 연비(mpg)를 예측하는 최적 모델을 만들고 RMSE로 평가 결과를 구하시오.

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np

data=pd.read_csv('data/mtcars.csv',index_col=0)
data.head()
```

```python
y=data[['mpg']]
#X=data[data.columns[1:]]
X=data[['cyl','disp','hp','drat','wt','vs','am','carb']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, train_size=0.7)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X_train)
X_scaled_train=scaler.transform(X_train)
X_scaled_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_scaled_train, y_train)
model.score(X_scaled_train, y_train)
```

```python
pred_test=model.predict(X_scaled_test)
model.score(X_scaled_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error

MSE=mean_squared_error(y_test, pred_test)
np.sqrt(MSE)
```

```python
from sklearn.svm import SVR
model=SVR(C=10, kernel='linear')
model.fit(X_scaled_train, y_train)
model.score(X_scaled_train, y_train)
```

```python
pred_test=model.predict(X_scaled_test)
model.score(X_scaled_test, y_test)
```

```python
MSE=mean_squared_error(y_test, pred_test)
result=np.sqrt(MSE)
print(result)
```
