4. 매출의 이상값을 제외한 데이터를 훈련 데이터로 선정할 때 Age의 표준편차를 구하시오.

```python
data=pd.read_csv('data/Carseats.csv')
data.head()
```

```python
mean=data.Sales.mean()
std=data.Sales.std()
upper=mean+1.5*std
lower=mean-1.5*std
```

```python
result=data[(data.Sales<upper)&(data.Sales>lower)].Age.std()
print(result)
```

5. Luggage.room의 결측값을 중앙값으로 변환한 후 변환 전, 후 평균의 차이를 구하시오.

```python
data=pd.read_csv('data/Cars93.csv', index_col=0)
data.head()
```

```python
before_mean=data['Luggage.room'].mean()
```

```python
data['Luggage.room'].fillna(data['Luggage.room'].median(), inplace=True)
```

```python
after_mean=data['Luggage.room'].mean()
```

```python
import numpy as np
result=np.abs(before_mean-after_mean)
print(result)
```

6. 연령이 20대인 확진자의 평균과 50대인 확진자의 평균의 차이를 구하시오.

```python
data=pd.read_csv('data/TimeAge.csv')
data.head()
```

```python
Twentys=data.groupby('age')['confirmed'].mean()['20s']
Fifties=data.groupby('age')['confirmed'].mean()['50s']
result=np.abs(Twentys-Fifties)
print(result)
```
