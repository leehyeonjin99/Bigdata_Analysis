```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')
plt.style.use('seaborn-deep')

import matplotlib.cm
cmap=matplotlib.cm.get_cmap('plasma')

from sklearn.cluster import KMeans

data=pd.read_csv('../data/Mall_Customers.csv', encoding='utf-8')
X=data[data.columns[3:]]
X.head()
```

```python
#       
wcss=[]

for i in range(1,21):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit_transform(X)
    wcss.append(kmeans.inertia_) # kmeans.inertia_ :    
    
wcss
```

```python
plt.figure()
plt.plot(range(1,21),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
k=5
kmeans=KMeans(n_clusters=k)
y_kmeans=kmeans.fit_predict(X) # fit_predict :   
y_kmeans
```

```python
Group_cluster=pd.DataFrame(y_kmeans)
Group_cluster.columns=['Group']
full_data=pd.concat([data, Group_cluster], axis=1)
full_data
```

```python
kmeans_pred=KMeans(n_clusters=k, random_state=42).fit(X)
kmeans_pred.cluster_centers_
```

```python
#        ?
kmeans_pred.predict([[100,50],[30,80]])
```

```python
labels=[('Cluster'+str(i+1)) for i in range(k)]
labels
```

```python
X=np.array(X)
plt.figure()
for i in range(k):
    plt.scatter(x=X[y_kmeans==i, 0], y=X[y_kmeans==i, 1],s=20, c=cmap(i/k), label=labels[i])
```

```python
#   
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
          s=100, c='black', label='Centroids', marker='X')
plt.xlabel('Income')
plt.ylabel('Spend')
plt.title('Kmeans cluster plot')
plt.legend()
plt.show()
```

## k-mean Clustering

```python
from scipy.spatial.distance import cdist, pdist
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

iris=pd.read_csv('../data/iris.csv')
iris.head()
```

```python
x_iris=iris.drop('class',axis=1)
y_iris=iris['class']
x_iris.head()
```

```python
y_iris.head()
```

```python
x_iris.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_iris)
X_scaled=scaler.transform(x_iris)
pd.DataFrame(X_scaled).head()
```

```python
K=range(1,10)
KM=[KMeans(n_clusters=k).fit(X_scaled) for k in K]
centroids=[k.cluster_centers_ for k in KM]
```

```python
D_k=[cdist(x_iris, centrds, 'euclidean') for centrds in centroids]
D_k
```

```python
cidx=[np.argmin(D, axis=1) for D in D_k]
dist=[np.min(D, axis=1) for D in D_k]
avgWithinSS=[sum(d)/X_scaled.shape[0] for d in dist]
```

```python
# TOtal with-in sum of square
wcss=[sum(d**2) for d in dist]
tss=sum(pdist(X_scaled)**2/X_scaled.shape[0])
bss=tss-wcss
```

```python
# elbow curve - Avg. within-cluster sum of squares
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(K, avgWithinSS, 'b*-')
plt.grid(True)
plt.xlabel('Number of clusters')
plt.ylabel('Average within-cluster sum of squares')
```

```python
# elbow curve - percentage of variance explained
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(K, bss/tss*100, 'b*-')
plt.grid(True)
plt.xlabel('Number of clusters')
plt.ylabel('Percentage of variance explained')
```

```python
#   eigenvectors & eigenvalues
import numpy as np
w, v =np.linalg.eig(np.array([[0.91335,0.75969],[0.75969, 0.69702]]))
print(f"Eigen Values : \n{w}\nEigen Vectors : \n{v}")
```

```python
k_means_fit=KMeans(n_clusters=4, max_iter=300)
k_means_fit.fit(X_scaled)
```

```python
k_means_fit.cluster_centers_
```

```python
# 3    4     
print('KMeans Clustering - Confusion Matrix\n\n',
     pd.crosstab(y_iris, k_means_fit.labels_, rownames=['Actuall'], colnames=['Predicted']))
```

```python
# Silhouette_score 
print(f"Silhouette- score : {silhouette_score(x_iris, k_means_fit.labels_, metric='euclidean'):.3f}")
```

```python
for k in range(2, 10):
    k_means_fit=KMeans(n_clusters=k, max_iter=100)
    k_means_fit.fit(x_iris)
    print(f"For K value {k}, Silhouette- score : {silhouette_score(x_iris, k_means_fit.labels_, metric='euclidean'):.3f}")
```
